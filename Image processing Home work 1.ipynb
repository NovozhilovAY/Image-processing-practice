Задание: построить графики функций
* sin(x) / x
* f(x) = ($x^3$ - $6x^2$ + 3x) / |x|
* $2^{sin(x)}$ и $2^x$ на одном графике
* f(x) = $x^2$ - 2x - 4 - |$x^2$ + x - 2|
* f(x) = 1 / (x-2) + 1
* y = $\sqrt{x}$$\sqrt{x-1}$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-100,100.1, 0.1)
plt.plot(x, np.sin(x) / x)
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin(x) / x")
plt.show
```

```python
x = np.arange(-100,100.1, 0.1)
plt.plot(x, (x**3 - 6*x**2 + 3 * x) / np.abs(x))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("($x^3$ - $6x^2$ + 3x) / |x|")
plt.show
```

```python
x = np.arange(0,5.01, 0.01)
plt.plot(x,2**x,x,x**(np.sin(x)))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("$2^{sin(x)}$, $2^x$")
plt.show
```

```python
x = np.arange(-10,10.1, 0.1)
plt.plot(x,x**2 - 2*x - 4 - np.abs(x**2 + x - 2))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x) = $x^2$ - 2x - 4 - |$x^2$ + x - 2|")
plt.show
```

```python
x1 = np.arange(2.1,6.1, 0.1)
x2 = np.arange(-3,1.9,0.1)
plt.plot(x1,(1 / (x1 - 2) + 1),"-r",x2,(1 / (x2 - 2)),"-r")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x) = 1/(x-2) + 1")

plt.show
```

```python
x = np.arange(0,1, 0.01)
plt.plot(x,np.sqrt(x)*np.sqrt(1-x))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x) = $\sqrt{x}$ $\sqrt{x-1}$")

plt.show
```
